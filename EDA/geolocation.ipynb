```python
import pandas as pd
```

```python
df =pd.read_csv('/content/jogja_fix_1.csv')
```

```python
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim(user_agent="my_geocoder",timeout=10)

# Fungsi untuk mendapatkan koordinat dari kecamatan di kota Yogyakarta
def get_coordinates(alamat):
    viewbox = [110.3600, -7.9300, 110.4600, -7.7900]
    full_address = f"{alamat}, Yogyakarta, Indonesia"
    location = geolocator.geocode(full_address)

    if location:
        return location.latitude, location.longitude
    else:
        return None, None
df[['latitude', 'longitude']] = df['alamat'].apply(get_coordinates).apply(pd.Series)
df.head()
```
