```python
fitur = [
    'sistem_alarm', 'gym', 'internet_broadband/wifi', 'tv_kabel', 'pemanas_ruangan','ac',
    'air_panas', 'telepon', 'tv', 'kitchen_set', 'garasi',
    'secure_parking', 'kolam_renang', 'lapangan_tenis', 'balkon', 'dek','halaman_terbuka',
    'area_hiburan_outdoor', 'pagar_penuh', 'taman', 'keamanan_24_jam', 'taman_bermain_anak'
]


for col in fitur:
  print(col)
  print(df[col].value_counts())
  print("-------")
```

```python
import numpy as np

total_fitur = len(fitur)
subplot_rows = int(np.ceil(total_fitur / 5))
subplot_cols = min(5, total_fitur)

fig, axes = plt.subplots(subplot_rows, subplot_cols, figsize=(15, 2 * subplot_rows))
fig.suptitle('Proporsi Fitur Lain', y=1.02)

# Flatten axes jika lebih dari satu baris
axes = axes.flatten()

for i, ax in enumerate(axes):
    if i < total_fitur:
        column = fitur[i]
        category_counts = df[column].value_counts()

        ax.bar(category_counts.index,
               category_counts.values,
               tick_label=category_counts.index,
               color='skyblue')  # Ganti warna sesuai keinginan

        ax.set_title(f'{column}')
plt.tight_layout()
plt.show()
```

```python
loc_mean = df.groupby('alamat')['harga'].mean().reset_index().sort_values(by='harga',ascending=True)
```

```python
plt.figure(figsize=(10,8))
plt.barh(loc_mean['alamat'],loc_mean['harga'])
plt.title('Rata-rata Harga')
plt.xlabel('harga rumah')
plt.ylabel('')
plt.grid(axis='x')
labels,location = plt.xticks()
plt.tight_layout()
plt.show()
```
